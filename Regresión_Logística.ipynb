```python
# Librerías necesarias
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Cargar los datos
df = pd.read_csv("C:\\Users\\ACER\\Downloads\\archive\\heart_cleveland_upload.csv")  # Reemplaza si la ruta cambia

# Visualizar las primeras filas
print(df.head())

# Verificar nombres de columnas
print("Columnas del dataset:", df.columns)

# Verificar si hay valores nulos
print("\nValores nulos por columna:\n", df.isnull().sum())

# Convertir nombres de columnas a minúsculas (opcional, buena práctica)
df.columns = df.columns.str.strip().str.lower()

# Variables predictoras (X) y variable objetivo (y)
X = df.drop("condition", axis=1)
y = df["condition"]

# Estandarizar las variables (muy útil para modelos como regresión logística)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir en conjunto de entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

print("\nTamaños de los datasets:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
# Selección simple usando correlación
correlation = df.corr()['quality'].drop('quality')
top_features = correlation[abs(correlation) > 0.1].index.tolist()
print("Características seleccionadas:", top_features)

X_selected = df[top_features]
```

```python
# Selección de características usando la correlación con la variable objetivo 'condition'
correlation = df.corr(numeric_only=True)['condition'].drop('condition')
top_features = correlation[abs(correlation) > 0.1].index.tolist()

print("Características seleccionadas por correlación (>0.1):", top_features)

# Crear subconjunto del dataset con las variables seleccionadas
X_selected = df[top_features]
y = df['condition']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# Crear el modelo de Regresión Logística
logreg = LogisticRegression(max_iter=1000, solver='lbfgs', multi_class='auto')

# Entrenar el modelo
logreg.fit(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression

# Crear el modelo de Regresión Logística sin el parámetro multi_class
logreg = LogisticRegression(max_iter=1000, solver='lbfgs')

# Entrenar el modelo
logreg.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Predecir en el conjunto de prueba
y_pred = logreg.predict(X_test)

# Métricas
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Reporte de clasificación:\n", classification_report(y_test, y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Matriz de Confusión - Regresión Logística")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()
```
